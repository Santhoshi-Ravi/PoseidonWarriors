```python
from newsapi import NewsApiClient
from urllib.request import urlopen
from bs4 import BeautifulSoup
from summarizer import Summarizer, TransformerSummarizer
```

```python
newsapi = NewsApiClient(api_key='haha, you wish')
actual_query = 'shark+tank OR shark OR  Blahaj'
all_articles = newsapi.get_everything(q='shark OR  Blahaj',from_param='2021-08-27',to='2021-08-27',language='en')
```

```python
bert_model = Summarizer()
```

```python
for key,value in all_articles.items():
  if key=='articles':
    list_of_articles = value
count = 0
link_list = {}
for article in list_of_articles:
  ct = str(count)
  to_open_url = article['url']
  link_list[ct] = {}
  link_list[ct]['url']= to_open_url
  link_list[ct]['title']= article['title']
  try:
    page = urlopen(to_open_url)
  except:
    pass
  soup = BeautifulSoup(page,'html.parser')
  para = ''
  for lines in soup.find_all('p'):
    para += lines.get_text()
    link_list[ct]['raw text'] = para

  count += 1
```

```python
s = []
count_of_stories = 0
for i in link_list.keys():
  if 'raw text' in link_list[i].keys() and count_of_stories!=11:
    mydict={}
    mydict['url'] = link_list[i]['url']
    mydict['title'] = link_list[i]['title']
    body = link_list[i]['raw text']
    bert_summary = ''.join(bert_model(body))
    mydict['summary'] = bert_summary
    s.append(mydict)
    count_of_stories += 1
    mydict = {}
    
```
